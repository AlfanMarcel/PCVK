```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print(' Mengubah tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness=int(input('masukkan nilai kecerahan : '))
except ValueError:
  print('error, not a number')

original =cv2.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')
brightness_image = np.zeros(original.shape,original.dtype)

for y in range(original.shape[0]):
    for x in range(original.shape[1]):
        for c in range(original.shape[2]):
            brightness_image[y,x,c]=np.clip(original[y,x,c] + brightness,0,255)

final_frame=cv2.hconcat ((original,brightness_image))
cv2_imshow(final_frame)
```

1. Implementasikan inverse citra pada Google Colaboratory menggunakan formula yang
terdapat pada bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:

```python
image =cv2.imread('/content/drive/MyDrive/PCVK/Images/ktp.jpeg')
inverse = np.zeros(image.shape,image.dtype)

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            inverse[y,x,c]=np.clip(255 - image[y,x,c] ,0 ,255)

final_frame=cv2.hconcat ((image,inverse))
cv2_imshow(final_frame)

# menggunakan library untuk invert gambar
# inverted_image = cv2.bitwise_not(image)
```

2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga
menghasilkan keluaran seperti berikut:

```python
print(' Mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------------------')
try:
  brightness=int(input('masukkan tingkat kecerahan [0-255] : '))
  contrast=int(input('masukkan kontrast [1.0-3.0]        : '))
except ValueError:
  print('error, not a number')

original =cv2.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')
contrast_image = np.zeros(original.shape,original.dtype)

for y in range(original.shape[0]):
    for x in range(original.shape[1]):
        for c in range(original.shape[2]):
            contrast_image[y,x,c]=np.clip(contrast * original[y,x,c] + brightness,0,255)

final_frame=cv2.hconcat ((original,contrast_image))
cv2_imshow(final_frame)

# menggunakan library untuk change contrast dann brightness
# editedImage=cv2.convertScaleAbs(original, alpha=contrast, beta=brightness)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log, sehingga menghasilkan keluaran seperti berikut:

```python
print(' Mengubah tingkat kecerahan citra dengan Transformasi Log ')
print('----------------------------------------------')
try:
  brightness=int(input('masukkan nilai kecerahan : '))
except ValueError:
  print('error, not a number')

original =cv2.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')
brightnesslog_image = np.zeros(original.shape,original.dtype)

for y in range(original.shape[0]):
    for x in range(original.shape[1]):
        for c in range(original.shape[2]):
            brightnesslog_image[y,x,c]=np.clip(brightness * np.log(1+ original[y,x,c]),0,255)


final_frame=cv2.hconcat ((original,brightnesslog_image))
cv2_imshow(final_frame)
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness,
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada
bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:

a. Averaging

```python
image =cv2.imread('/content/drive/MyDrive/PCVK/Images/ktp.jpeg')
grayaverage_image = np.zeros(image.shape,image.dtype)

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
          red = image[y,x,2] # red channel
          green = image[y,x,1] # green channel
          blue = image[y,x,0] # blue channel
          average = ((red/3) + (green/3) + (blue/3))
          grayaverage_image[y,x,c] = np.clip(average,0,255)

final_frame=cv2.hconcat ((image, grayaverage_image))
cv2_imshow(final_frame)
```

b. Lightness

```python
image =cv2.imread('/content/drive/MyDrive/PCVK/Images/ktp.jpeg')
graylightness_image = np.zeros(image.shape,image.dtype)

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
          red = image[y,x,2] # red channel
          green = image[y,x,1] # green channel
          blue = image[y,x,0] # blue channel
          lightness = ((max(red,green,blue)/2) + (min(red,green,blue)/2))
          graylightness_image[y,x,c] = np.clip(lightness,0,255)

final_frame=cv2.hconcat ((image, graylightness_image))
cv2_imshow(final_frame)
```

c. Luminance

```python
image =cv2.imread('/content/drive/MyDrive/PCVK/Images/ktp.jpeg')
grayluminance_image = np.zeros(image.shape,image.dtype)

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
          red = image[y,x,2] # red channel
          green = image[y,x,1] # green channel
          blue = image[y,x,0] # blue channel
          luminance = ((0.21*red) + (0.07*blue) +(0.72*green))
          grayluminance_image[y,x,c] = np.clip(luminance,0,255)

final_frame=cv2.hconcat ((image, grayluminance_image))
cv2_imshow(final_frame)
```

5. Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale. Misal,
tampilkan warna biru pada citra masukan dan ubah bagian lain yang tidak berwarna
biru menjadi grayscale seperti pada contoh berikut:

```python
image =cv2.imread('/content/drive/MyDrive/PCVK/Images/ktp.jpeg')
last_image = np.zeros(image.shape,image.dtype)

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
          red = image[y,x,2] # red channel
          green = image[y,x,1] # green channel
          blue = image[y,x,0] # blue channel
          luminance = ((0.21*red) + (0.07*blue) +(0.72*green))

          last_image[y,x,0] = blue # blue channel
          last_image[y,x,2] = np.clip(luminance,0,255) # red channel
          last_image[y,x,1] = np.clip(luminance,0,255) # green channel

final_frame=cv2.hconcat ((image, last_image))
cv2_imshow(final_frame)
```
